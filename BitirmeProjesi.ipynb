**Apriori Algoritmasında Satır ve Sütunların Hazırlanması**

```python
my_ds = df.groupby(['Order ID', 'Item ID'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Order ID')
def binary_table(x):
  if(x>0):
    return 1
  else:
    return 0
my_ds = my_ds.applymap(lambda x:binary_table(x))
my_ds #ilgili satırdaki Order ID'de bulunan Item ID'ler 1, bulunmayanlar 0 olarak tabloda yer almaktadır
```

**Sıklıkla Kullanılan Itemset'lerin Bulunması ve Support Değerleri**

```python
import time
ap_start = time.time()
from mlxtend.frequent_patterns import apriori, association_rules #apriori kütüphanesinin eklenmesi
frequency = apriori(my_ds, min_support=0.01, use_colnames=True)
ap_end = time.time()
frequency
#my1=frequency[frequency["itemsets"]=={265}]
```

```python
ap_time=ap_end-ap_start
ap_time #apriori algoritması çalışma süresi
```

**Apriori Algoritması Birliktelik Kuralları**

```python
rules_ap_start = time.time()
rules = association_rules(frequency, metric="support", min_threshold=0.01)
rules_ap_end = time.time()
rules
```

```python
rules_ap_time=rules_ap_end-rules_ap_start
rules_ap_time #apriori algoritması kural oluşturma süresi
```

**Müşteri Seçiminin Filtrelenip Confidence'a göre Dataframe'in Sıralanması**

```python
sorted_antecedents=rules.loc[rules['antecedents'] == {customer_choice(choice)}].sort_values(by="confidence",ascending=False)
if sorted_antecedents.index.empty:
  raise Exception("İstenen Item ID için itemset oluşturulamadı...")
sorted_antecedents
```

**Müşterinin Satın Alması Tahmin Edilen Ürünler**

```python
predictions=sorted_antecedents["consequents"].head(5) #confidence değeri en yüksek ilk 5 siparişin getirilmesi
pred_list=[list(i) for i in predictions]
#print(pred_list)
pred_list1 = [item for sublist in pred_list for item in sublist] #nested list normal list'e dönüştürüldü
def unique(list1):
    x = np.array(list1)
    return np.unique(x)
pred_list1=unique(pred_list1) #unique olmayan elemanların listeden çıkarılması
result=[]
for j in range(0,len(pred_list1)):
 result.append(get_item(df,pred_list1[j])) #çıkan tahmin sonucunu boş listeye aktarıyoruz
#print(result)
#temp_budget=customer_budget
print("Müşterinin satın alması tahmin edilen ürün(ler) ve bütçesi dahilinde maksimum satın alınabilecek adet:")
for k in range(0,len(result)):
    max_item_quantity=customer_budget/(df[df['Item Name']==result[k]]['Product Price'].values[0])
    import math
    max_item_quantity=math.floor(max_item_quantity)
    print(f"{k+1}-",result[k],f"(max {max_item_quantity} adet)",)
```

**Stok Artırılmasının Gerekli Olduğu Durumların Tespiti (Belirli Tarih için)**

```python
switch=1
for i in range(0,100):
    while switch==1:
      try:
        year,month,day = input("Tarih seçiniz: (YYYY MM DD formatında)\n").split()
        year=int(year)
        month=int(month)
        day=int(day)
        pred_list_df=df[df['Item ID'].isin(pred_list1)]
        date_filtered_df=pred_list_df[(pred_list_df['Date'] == pd.Timestamp(year,month,day))]
        if date_filtered_df.empty==False:
          switch=0
        else:
          print("İstenen tarihte herhangi bir veri bulunamadı...")
      except:
          print("Hata...")
          continue
      break
date_filtered_df
```

**Dataframe'de İstenen Tarih ve Saat Aralığındaki En Güncel Stok Sayısını Elde Etmek için Time Sütununu Sıralama İşlemi**

```python
sorted_time=date_filtered_df.loc[date_filtered_df['Item ID'] == date_filtered_df["Item ID"]].sort_values(by="Time",ascending=False)
sorted_time
```

**Stok Durum Kontrolü**

```python
for z in range(0,len(result)):
  try:
    max_item_quantity=customer_budget/(df[df['Item Name']==result[z]]['Product Price'].values[0])
    max_item_quantity=math.floor(max_item_quantity)
    #print(max_item_quantity)
    stock_control=(sorted_time[sorted_time['Item ID']==pred_list1[z]]['Total products'].values[0]-max_item_quantity)
    if stock_control<0:
      print(f"{result[z]} için {abs(stock_control)} adet stok artırılması tavsiye edilmektedir")
    else:
      print(f"{result[z]} için stok yeterlidir")
  except:
    pass
    #print("İstenilen tarih için stok bilgisi mevcut değil...")
```

**FP-Growth Algoritması**

```python
fp_start = time.time()
from mlxtend.frequent_patterns import fpgrowth
frequency_fp=fpgrowth(my_ds, min_support=0.01, use_colnames=True) #my_ds değişkenini apriori algoritmasında tanımladık, min_support'u 0.012 alabiliriz
#apriori algoritmasında satır ve sütunları binary şeklinde hazırlamıştık(Order ID'de bulunan Item ID'ler 1, bulunmayanlar 0)
fp_end = time.time()
frequency_fp
```

```python
fp_time=fp_end-fp_start
fp_time #fp-growth algoritması çalışma süresi
```

```python
from mlxtend.frequent_patterns import association_rules
rules_fp_start = time.time()
rules_fp=association_rules(frequency_fp, metric="support", min_threshold=0.01) #metric'i lift min_threshold'u 1 alabiliriz
rules_fp_end = time.time()
rules_fp
```

```python
rules_fp_time=rules_fp_end-rules_fp_start
rules_fp_time #fp-growth algoritması kural oluşturma süresi
```

```python
sorted_antecedents_fp=rules_fp.loc[rules_fp['antecedents'] == {customer_choice(choice)}].sort_values(by="confidence",ascending=False)
sorted_antecedents_fp
```

```python
predictions_fp=sorted_antecedents_fp["consequents"].head(5) #confidence değeri en yüksek ilk 5 siparişin getirilmesi
pred_list_fp=[list(i) for i in predictions_fp]
#print(pred_list_fp)
pred_list_fp1 = [item for sublist in pred_list_fp for item in sublist] #nested list normal list'e dönüştürüldü
def unique_fp(list1):
    x = np.array(list1)
    return np.unique(x)
pred_list_fp1=unique_fp(pred_list_fp1) #unique olmayan elemanların listeden çıkarılması
result_fp=[]
for j in range(0,len(pred_list_fp1)):
 result_fp.append(get_item(df,pred_list_fp1[j])) #çıkan tahmin sonucunu boş listeye aktarıyoruz
#print(result_fp)
#temp_budget=customer_budget
print("Müşterinin satın alması tahmin edilen ürün(ler) ve bütçesi dahilinde maksimum satın alınabilecek adet:")
for k in range(0,len(result_fp)):
    max_item_quantity=customer_budget/(df[df['Item Name']==result_fp[k]]['Product Price'].values[0])
    import math
    max_item_quantity=math.floor(max_item_quantity)
    print(f"{k+1}-",result_fp[k],f"(max {max_item_quantity} adet)",)
```

**Apriori ve FP-Growth Algoritmaları için Çalışma Sürelerinin Karşılaştırılması**

```python
print("İki algoritmanın çalışma süreleri:")
print(f"Apriori algoritması: {round(ap_time, 5)} saniye")
print(f"FP-Growth algoritması: {round(fp_time, 5)} saniye")
a=float("{:.2f}".format(ap_time/fp_time))
print(f"FP-Growth, Apriori'ye göre {a} kat daha hızlı çalışmaktadır")
```
