**Apriori AlgoritmasÄ±nda SatÄ±r ve SÃ¼tunlarÄ±n HazÄ±rlanmasÄ±**

```python
my_ds = df.groupby(['Order ID', 'Item ID'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Order ID')
def binary_table(x):
  if(x>0):
    return 1
  else:
    return 0
my_ds = my_ds.applymap(lambda x:binary_table(x))
my_ds #ilgili satÄ±rdaki Order ID'de bulunan Item ID'ler 1, bulunmayanlar 0 olarak tabloda yer almaktadÄ±r
```

**SÄ±klÄ±kla KullanÄ±lan Itemset'lerin BulunmasÄ± ve Support DeÄŸerleri**

```python
import time
ap_start = time.time()
from mlxtend.frequent_patterns import apriori, association_rules #apriori kÃ¼tÃ¼phanesinin eklenmesi
frequency = apriori(my_ds, min_support=0.01, use_colnames=True)
ap_end = time.time()
frequency
#my1=frequency[frequency["itemsets"]=={265}]
```

```python
ap_time=ap_end-ap_start
ap_time #apriori algoritmasÄ± Ã§alÄ±ÅŸma sÃ¼resi
```

**Apriori AlgoritmasÄ± Birliktelik KurallarÄ±**

```python
rules_ap_start = time.time()
rules = association_rules(frequency, metric="support", min_threshold=0.01)
rules_ap_end = time.time()
rules
```

```python
rules_ap_time=rules_ap_end-rules_ap_start
rules_ap_time #apriori algoritmasÄ± kural oluÅŸturma sÃ¼resi
```

**MÃ¼ÅŸteri SeÃ§iminin Filtrelenip Confidence'a gÃ¶re Dataframe'in SÄ±ralanmasÄ±**

```python
sorted_antecedents=rules.loc[rules['antecedents'] == {customer_choice(choice)}].sort_values(by="confidence",ascending=False)
if sorted_antecedents.index.empty:
  raise Exception("Ä°stenen Item ID iÃ§in itemset oluÅŸturulamadÄ±...")
sorted_antecedents
```

**MÃ¼ÅŸterinin SatÄ±n AlmasÄ± Tahmin Edilen ÃœrÃ¼nler**

```python
predictions=sorted_antecedents["consequents"].head(5) #confidence deÄŸeri en yÃ¼ksek ilk 5 sipariÅŸin getirilmesi
pred_list=[list(i) for i in predictions]
#print(pred_list)
pred_list1 = [item for sublist in pred_list for item in sublist] #nested list normal list'e dÃ¶nÃ¼ÅŸtÃ¼rÃ¼ldÃ¼
def unique(list1):
    x = np.array(list1)
    return np.unique(x)
pred_list1=unique(pred_list1) #unique olmayan elemanlarÄ±n listeden Ã§Ä±karÄ±lmasÄ±
result=[]
for j in range(0,len(pred_list1)):
 result.append(get_item(df,pred_list1[j])) #Ã§Ä±kan tahmin sonucunu boÅŸ listeye aktarÄ±yoruz
#print(result)
#temp_budget=customer_budget
print("MÃ¼ÅŸterinin satÄ±n almasÄ± tahmin edilen Ã¼rÃ¼n(ler) ve bÃ¼tÃ§esi dahilinde maksimum satÄ±n alÄ±nabilecek adet:")
for k in range(0,len(result)):
    max_item_quantity=customer_budget/(df[df['Item Name']==result[k]]['Product Price'].values[0])
    import math
    max_item_quantity=math.floor(max_item_quantity)
    print(f"{k+1}-",result[k],f"(max {max_item_quantity} adet)",)
```

**Stok ArtÄ±rÄ±lmasÄ±nÄ±n Gerekli OlduÄŸu DurumlarÄ±n Tespiti (Belirli Tarih iÃ§in)**

```python
switch=1
for i in range(0,100):
    while switch==1:
      try:
        year,month,day = input("Tarih seÃ§iniz: (YYYY MM DD formatÄ±nda)\n").split()
        year=int(year)
        month=int(month)
        day=int(day)
        pred_list_df=df[df['Item ID'].isin(pred_list1)]
        date_filtered_df=pred_list_df[(pred_list_df['Date'] == pd.Timestamp(year,month,day))]
        if date_filtered_df.empty==False:
          switch=0
        else:
          print("Ä°stenen tarihte herhangi bir veri bulunamadÄ±...")
      except:
          print("Hata...")
          continue
      break
date_filtered_df
```

**Dataframe'de Ä°stenen Tarih ve Saat AralÄ±ÄŸÄ±ndaki En GÃ¼ncel Stok SayÄ±sÄ±nÄ± Elde Etmek iÃ§in Time SÃ¼tununu SÄ±ralama Ä°ÅŸlemi**

```python
sorted_time=date_filtered_df.loc[date_filtered_df['Item ID'] == date_filtered_df["Item ID"]].sort_values(by="Time",ascending=False)
sorted_time
```

```python
for z in range(0,len(result)):
  try:
    max_item_quantity=customer_budget/(df[df['Item Name']==result[z]]['Product Price'].values[0])
    max_item_quantity=math.floor(max_item_quantity)
    #print(max_item_quantity)
    stock_control=(sorted_time[sorted_time['Item ID']==pred_list1[z]]['Total products'].values[0]-max_item_quantity)
    if stock_control<0:
      print(f"{result[z]} iÃ§in {abs(stock_control)} adet stok artÄ±rÄ±lmasÄ± tavsiye edilmektedir")
    else:
      print(f"{result[z]} iÃ§in stok yeterlidir")
  except:
    pass
    #print("Ä°stenilen tarih iÃ§in stok bilgisi mevcut deÄŸil...")
```

**FP-Growth AlgoritmasÄ±**

```python
fp_start = time.time()
from mlxtend.frequent_patterns import fpgrowth
frequency_fp=fpgrowth(my_ds, min_support=0.01, use_colnames=True) #my_ds deÄŸiÅŸkenini apriori algoritmasÄ±nda tanÄ±mladÄ±k, min_support'u 0.012 alabiliriz
#apriori algoritmasÄ±nda satÄ±r ve sÃ¼tunlarÄ± binary ÅŸeklinde hazÄ±rlamÄ±ÅŸtÄ±k(Order ID'de bulunan Item ID'ler 1, bulunmayanlar 0)
fp_end = time.time()
frequency_fp
```

```python
fp_time=fp_end-fp_start
fp_time #fp-growth algoritmasÄ± Ã§alÄ±ÅŸma sÃ¼resi
```

```python
from mlxtend.frequent_patterns import association_rules
rules_fp_start = time.time()
rules_fp=association_rules(frequency_fp, metric="support", min_threshold=0.01) #metric'i lift min_threshold'u 1 alabiliriz
rules_fp_end = time.time()
rules_fp
```

```python
rules_fp_time=rules_fp_end-rules_fp_start
rules_fp_time #fp-growth algoritmasÄ± kural oluÅŸturma sÃ¼resi
```

```python
sorted_antecedents_fp=rules_fp.loc[rules_fp['antecedents'] == {customer_choice(choice)}].sort_values(by="confidence",ascending=False)
sorted_antecedents_fp
```

```python
predictions_fp=sorted_antecedents_fp["consequents"].head(5) #confidence deÄŸeri en yÃ¼ksek ilk 5 sipariÅŸin getirilmesi
pred_list_fp=[list(i) for i in predictions_fp]
#print(pred_list_fp)
pred_list_fp1 = [item for sublist in pred_list_fp for item in sublist] #nested list normal list'e dÃ¶nÃ¼ÅŸtÃ¼rÃ¼ldÃ¼
def unique_fp(list1):
    x = np.array(list1)
    return np.unique(x)
pred_list_fp1=unique_fp(pred_list_fp1) #unique olmayan elemanlarÄ±n listeden Ã§Ä±karÄ±lmasÄ±
result_fp=[]
for j in range(0,len(pred_list_fp1)):
 result_fp.append(get_item(df,pred_list_fp1[j])) #Ã§Ä±kan tahmin sonucunu boÅŸ listeye aktarÄ±yoruz
#print(result_fp)
#temp_budget=customer_budget
print("MÃ¼ÅŸterinin satÄ±n almasÄ± tahmin edilen Ã¼rÃ¼n(ler) ve bÃ¼tÃ§esi dahilinde maksimum satÄ±n alÄ±nabilecek adet:")
for k in range(0,len(result_fp)):
    max_item_quantity=customer_budget/(df[df['Item Name']==result_fp[k]]['Product Price'].values[0])
    import math
    max_item_quantity=math.floor(max_item_quantity)
    print(f"{k+1}-",result_fp[k],f"(max {max_item_quantity} adet)",)
```

**Apriori ve FP-Growth AlgoritmalarÄ± iÃ§in Ã‡alÄ±ÅŸma SÃ¼relerinin KarÅŸÄ±laÅŸtÄ±rÄ±lmasÄ±**

```python
print("Ä°ki algoritmanÄ±n Ã§alÄ±ÅŸma sÃ¼releri:")
print(f"Apriori algoritmasÄ±: {round(ap_time, 5)} saniye")
print(f"FP-Growth algoritmasÄ±: {round(fp_time, 5)} saniye")
a=float("{:.2f}".format(ap_time/fp_time))
print(f"FP-Growth, Apriori'ye gÃ¶re {a} kat daha hÄ±zlÄ± Ã§alÄ±ÅŸmaktadÄ±r")
```
